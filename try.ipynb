```python
import random


def swap_letters(mapping):
    keys = list(mapping.keys())
    values = list(mapping.values())

    while True:
        key1, key2 = random.sample(keys, 2)
        if mapping[key1] != " " and mapping[key2] != " ":
            break

    value1, value2 = mapping[key1], mapping[key2]

    modified_mapping = dict(mapping)
    modified_mapping[key1] = value2
    modified_mapping[key2] = value1

    return modified_mapping


alphabet_dict = {
    "a": "a",
    "b": "b",
    "c": "c",
    "d": "d",
    "e": "e",
    "f": "f",
    "g": "g",
    "h": "h",
    "i": "i",
    "j": "j",
    "k": "k",
    "l": "l",
    "m": "m",
    "n": "n",
    "o": "o",
    "p": "p",
    "q": "q",
    "r": "r",
    "s": "s",
    "t": "t",
    "u": "u",
    "v": "v",
    "w": "w",
    "x": "x",
    "y": "y",
    "z": "z",
    " ": " ",
}
```
